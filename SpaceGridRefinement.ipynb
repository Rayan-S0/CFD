```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos, pi, log
N = np.array([10,100,1000,10000])
def BuildMatrix(N):  
    M = np.zeros([N,N])
    for i in range(N-1):
        M[i,i+1] = 1
        M[i,i] = -1
        M[N-1,0] = 1
        M[N-1,N-1] = -1
    return M
def Buildphis(n):
    h = 2/n
    x = np.linspace(-1, 1-h, num=n)
    s = np.cos(x)
    return s

def Buildphiprims(n):
    h = 2/n
    x = np.linspace(-1, 1-h, num=n)
    s = -np.sin(x)
    return s
for n in N:
    h = 2/n
    A = (1/h)*BuildMatrix(n)
    B = Buildphis(n)
    S = Buildphiprims(n)
    U = A.dot(B)
    error = abs(S-U)
    lm = np.average(error)
    one = plt.scatter(log(h),log(lm), color = 'red')
plt.plot([log(1/10),log(1/10000)], [log(1/10),log(1/10000)],color = 'green')
plt.title('Grid refinement for fist order forward difference')
plt.show()
```
