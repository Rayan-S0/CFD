```python
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
import numpy as np
from time import sleep
from math import pi, cos , sin
from IPython import display
#%matplotlib notebook

#user defined variables
N = 150 #number of grid points
cfl = 0.1 #v
endtime = 1 #simulation time
tview = 5 #when u want to see the simualation
space_dis = 'cos' # or 'sin' or 'e'
space_dis = 'fd1' # or 'cd2' or 'ccd4'
time_dis = 'euler' # or 'cn' or 'rk4'

#Preproccecsing definitions
a = -1
l = 2*pi
h = l/N
delT = abs(cfl*h/a)
StepN = round(endtime/delT)
x = np.linspace(0, l-h, num=N)


def Buildphis(n):
    s = np.cos(pi*x)
    return s


def derivate(N,s):
    M = np.zeros([N,N])
    for i in range(N-1):
        M[i,i+1] = 1
        M[i+1,i] = -1
        M[0,N-1] = -1
        M[N-1,0] = 1
    M = (1/(2*h))*M
    D = M.dot(s)
    return D

S = np.zeros([StepN,N])
S[0,:] = Buildphis(N)
for i in range(StepN-1):
    S[i+1,:] = S[i,:] +delT*(-a)*derivate(N,S[i,:])
    
    
    

```
